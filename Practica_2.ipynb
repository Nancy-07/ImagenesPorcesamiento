```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img=Image.open('morgana.jpg')
img.show()
arrimg=np.array(img)
```

```python
newaimg = np.copy(arrimg)
def sumaimg (arr , a):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = arr[c][k][0] + a
      arr[c][k][1] = arr[c][k][1] + a
      arr[c][k][2] = arr[c][k][2] + a
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
  plt.imshow(newaimg)
  plt.show()
sumaimg(newaimg,50)
```

```python
newimg = np.copy(arrimg)
def restimg (arr , a):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = arr[c][k][0] - a
      arr[c][k][1] = arr[c][k][1] - a
      arr[c][k][2] = arr[c][k][2] - a
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
     
  plt.imshow(newimg)
  plt.show()
```

```python
newaaimg = np.copy(arrimg)
def divimg (arr , a):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = arr[c][k][0] / a
      arr[c][k][1] = arr[c][k][1] / a
      arr[c][k][2] = arr[c][k][2] / a
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
  plt.imshow(newaaimg)
  plt.show()
```

```python
newaaaimg=np.copy(arrimg)
def porimg (arr , a):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = arr[c][k][0] * a
      arr[c][k][1] = arr[c][k][1] * a
      arr[c][k][2] = arr[c][k][2] * a
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
  plt.imshow(newaaaimg)
  plt.show()
```

```python
newaaaaimg=np.copy(arrimg)
def gamaimg (arr ,a):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = round(255*((arr[c][k][0]/255)**(1/a)) )
      arr[c][k][1] = round(255*((arr[c][k][1]/255)**(1/a)) )
      arr[c][k][2] = round(255*((arr[c][k][2]/255)**(1/a)) )
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
  plt.imshow(newaaaaimg)
  plt.show()
```

```python
newaaaaaimg=np.copy(arrimg)
def inverimg (arr):
  for k in range(img.size[0]-1):
    for c in range(img.size[1]-1):
      arr[c][k][0] = 255 - arr[c][k][0] 
      arr[c][k][1] = 255- arr[c][k][1] 
      arr[c][k][2] = 255-arr[c][k][2] 
      if (arr[c][k][0]>=255):
        arr[c][k][0] =255
      elif(arr[c][k][1]>=255):
        arr[c][k][1] =255
      elif(arr[c][k][2]>=255):
        arr[c][k][2] =255
      if (arr[c][k][0]<=0):
        arr[c][k][0] = 0
      elif(arr[c][k][1])<=0:
        arr[c][k][1] = 0
      elif(arr[c][k][2])<=0:
        arr[c][k][2] = 0
  plt.imshow(newaaaaaimg)
  plt.show()
```

```python
arryec = np.copy(arrimg)
def ecualizacion(arryecu):
    wi = img.size[0]
    he = img.size[1]
    canalR=np.zeros(256)
    canalG=np.zeros(256)
    canalB=np.zeros(256)
    for k in range(img.size[0]-1):
        for c in range(img.size[1]-1):
            canalR[arryecu[c][k][0]] = canalR[arryecu[c][k][0]] + 1
            canalG[arryecu[c][k][1]] = canalG[arryecu[c][k][1]] + 1
            canalB[arryecu[c][k][2]] = canalB[arryecu[c][k][2]] + 1

    canalR=[y/(wi*he) for y in canalR]
    canalG=[y/(wi*he) for y in canalG]
    canalB=[y/(wi*he) for y in canalB]

    nuevosR = np.zeros(256)
    nuevosG = np.zeros(256)
    nuevosB = np.zeros(256)

    for i in range(256):
        nuevosR[i] = int(np.round(255*sum(canalR[0:i])))
        nuevosG[i] = int(np.round(255*sum(canalG[0:i])))
        nuevosB[i] = int(np.round(255*sum(canalB[0:i])))


    for k in range(img.size[0]-1):
        for c in range(img.size[1]-1):
            arryecu[c][k][0] = nuevosR[arryecu[c][k][0]]
            arryecu[c][k][1] = nuevosG[arryecu[c][k][1]]
            arryecu[c][k][2] = nuevosB[arryecu[c][k][2]]
            if (arryecu[c][k][0]>=255):
                arryecu[c][k][0] =255
            elif(arryecu[c][k][1]>=255):
                arryecu[c][k][1] =255
            elif(arryecu[c][k][2]>=255):
                arryecu[c][k][2] =255
            if (arryecu[c][k][0]<=0):
                arryecu[c][k][0] = 0
            elif(arryecu[c][k][1])<=0:
                arryecu[c][k][1] = 0
            elif(arryecu[c][k][2])<=0:
                arryecu[c][k][2] = 0
    plt.imshow(arryecu)
    plt.show()
```

```python
from PIL import Image

# matriz de valores de pxeles en escala de grises
pixels = [[0, 255, 0],
          [127, 0, 127],
          [255, 255, 255]]

# tamao de la imagen
width = len(pixels[0])
height = len(pixels)

# crear una imagen en escala de grises vaca
img = Image.new('L', (width, height))

# establecer valores de pxeles en la imagen
for y in range(height):
    for x in range(width):
        img.putpixel((x, y), pixels[y][x])

# guardar la imagen
img.save('output.png')
img.show()
```

```python
from PIL import Image
import numpy as np


img = Image.open('Avogado6.jpg')
width, height = img.size

img_translated = Image.new(img.mode, img.size)

# Definir la cantidad de traslacin en las coordenadas x e y
dx = 100
dy = 50

# Copiar cada pxel de la imagen original en la posicin trasladada en la nueva imagen
for x in range(width-dx):
    for y in range(height-dy):
        pixel = img.getpixel((x, y))
        img_translated.putpixel((x + dx, y + dy), pixel)

img_translated.save('imagen_trasladada.jpg')
img.show()
```

```python
import cv2
import numpy as np

# Cargar la imagen
img = cv2.imread('Avogado6.jpg')
# Definir la matriz de transformacin
M = np.float32([[1, 0, -100],
                [0, 1, -200]])

# Aplicar la traslacin
img_translated = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
# Mostrar la imagen trasladada
cv2.imshow('Imagen trasladada', img_translated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
