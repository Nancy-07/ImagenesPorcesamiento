```python
from PIL import Image, ImageDraw

# Crear una nueva imagen con fondo negro
image = Image.new("RGB", (400, 400), "black")
draw = ImageDraw.Draw(image)

# Definir el tamaño y los colores de los círculos
circle_size = 20
white_color = (255, 255, 255)
black_color = (0, 0, 0)

# Generar la matriz de ondas circulares
for i in range(20):
    for j in range(20):
        x = i * (circle_size + 10) + 10
        y = j * (circle_size + 10) + 10
        if (i + j) % 2 == 0:
            draw.ellipse([(x, y), (x + circle_size, y + circle_size)], fill=white_color)
        else:
            draw.ellipse([(x, y), (x + circle_size, y + circle_size)], fill=black_color)

# Mostrar la imagen generada
image.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones de la matriz y la imagen
width = 400
height = 400

# Parámetros de la onda de agua
amplitude = 10  # Amplitud de la onda
frequency = 5  # Frecuencia de la onda

# Crear la matriz de la onda de agua
x = np.linspace(0, width, width)
y = np.linspace(0, height, height)
X, Y = np.meshgrid(x, y)
Z = amplitude * np.sin(2 * np.pi * frequency * (X + Y) / width)

# Mostrar la matriz
print("Matriz de la onda de agua:")
print(Z)

# Mostrar la imagen de la onda de agua
plt.imshow(Z, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones de la imagen
width = 400
height = 400

# Parámetros de la onda de expansión
amplitude = 50  # Amplitud de la onda
center_x = width // 2  # Centro de la onda en el eje x
center_y = height // 2  # Centro de la onda en el eje y

# Crear la imagen de la onda de expansión
image = np.zeros((height, width), dtype=np.float32)

for x in range(width):
    for y in range(height):
        distance = np.sqrt((x - center_x)**2 + (y - center_y)**2)
        value = amplitude * np.exp(-distance / 50)  # Función de decaimiento exponencial
        image[y, x] = value

# Mostrar la imagen en escala de grises
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones de la imagen
width = 400
height = 600

# Parámetros del patrón de ondas circulares
radius = 200  # Radio de las ondas
frequency = 3  # Frecuencia de las ondas

# Crear la imagen
image = np.zeros((height, width), dtype=np.float32)

# Calcular el centro de la imagen
center_x = width // 2
center_y = height 

# Generar el patrón de ondas circulares
for x in range(width):
    for y in range(height):
        distance = np.sqrt((x - center_x) ** 2 + (y - center_y) ** 2)
        value = (np.sin(2 * np.pi * distance / radius * frequency) + 1) / 2
        image[y, x] = value

# Mostrar la imagen con el patrón de ondas circulares
plt.imshow(image, cmap='gray')
plt.show()
```
